### Grids with a loop
Given a canvas size we and a desired number of 



```python
from py5canvas import *
create_canvas(400, 400)
background(128)
rect_mode(CENTER)
no_fill()
n_rows, n_cols = 7, 5
x_step = width / n_cols
y_step = height / n_rows
stroke_weight(1)
for i in range(n_rows):
    for j in range(n_cols):
        pos = [x_step*j + x_step*0.5, y_step*i + y_step*0.5]
        circle(pos, 4)
        rect(pos, [x_step, y_step])
    
show()
```
